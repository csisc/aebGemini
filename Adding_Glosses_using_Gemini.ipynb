```python
import google.generativeai as genai
genai.configure(api_key="[GEMINI API KEY]")
for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
import time
model = genai.GenerativeModel("gemini-1.5-flash")


def check_relation_with_gemini(written, phonetic, pos_):
  print(written)
  response = model.generate_content(f"Could you provide the meaning of the Tunisian word '{written} {phonetic} ({pos_})'. Just provide the meaning. If there are two meanings or more, separate them by semi-colon.")
  generated_text = response.text
  print(generated_text)
  return generated_text


df["senses"] = ""  # Initialize a new column for Gemini validity

for index, row in df.iterrows():
  written = row["WrittenForm"]
  phonetic = row["PhoneticForm"]
  pos_ = row["PartOfSpeech"]
  try:
    is_valid = check_relation_with_gemini(written, phonetic, pos_)
  except AttributeError:
    is_valid = "UNKNOWN"
  time.sleep(7)
  df.at[index, "senses"] = is_valid

df.to_excel("lexicon.xlsx", index=False)
```
